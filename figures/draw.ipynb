```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set_theme(style="white")
```

```python
# 创建一个示例数据框
data = {'X': [1, 2, 3, 4, 5], 'Y': [5, 4, 3, 2, 1]}
df = pd.DataFrame(data)

# 绘制折线图
sns.lineplot(x='X', y='Y', data=df)
sns.despine()
plt.savefig("demo.svg", format='svg')
```
